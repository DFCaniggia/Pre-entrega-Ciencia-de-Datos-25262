```python
# Importamos los Datasets marketing, clientes, ventas.

print('\n*** Marketing ***\n')
df_marketing = pd.read_csv('/content/drive/MyDrive/datasets/marketing.csv')
df_marketing.info()

print('\n*** Clientes ***\n')
df_clientes = pd.read_csv('/content/drive/MyDrive/datasets/clientes.csv')
df_clientes.info()

print('\n*** Ventas ***\n')
df_ventas = pd.read_csv('/content/drive/MyDrive/datasets/ventas.csv')
df_ventas.info()
```

```python
# 2. Realizar un script básico que calcule las ventas mensuales utilizando
# variables y operadores.
# 3. Estructuras de Datos: Desarrollar un programa que almacene los datos de ventas (producto, precio, cantidad).
# Decidir si conviene utilizar diccionarios o listas.

# ==========================================================================
# ========== Pequeño script que limpia y nos da el total de ventas =========
# ==========================================================================

# Usamos la libreria csv
import csv

total_ventas = []

with open('/content/drive/MyDrive/datasets/ventas.csv', 'r', newline='', encoding='utf-8') as ventas_csv:

# Usamos readlines() para leer el encabezado y saber si estan las categorias producto, precio y cantidad.
    lector = csv.DictReader(ventas_csv)
    print(lector.fieldnames)

    total_fila: float



    for fila in lector:
        # Extraer y limpiar datos
        producto_ = fila['producto'].strip()
        precio_ = fila['precio'].replace('$', '').strip()
        cantidad_ = fila['cantidad'].strip()

        if precio_ == "":
            precio_limpio = 0.0 # No e guarda en la lista total_ventas.
        else:
            precio_limpio = float(precio_)
            cantidad_limpia = int(cantidad_)
            total_fila = precio_limpio * cantidad_limpia
            total_ventas.append(total_fila)


        print(f'Producto: {producto_}, Precio: {precio_limpio}, Cantidad: {cantidad_limpia}, Total: {total_fila:.2f}')


print(f'Suma total de ventas: {sum(total_ventas)}')


```

```python
# 4- Introducción a Pandas: realizar un análisis exploratorio inicial de los DataFrames.
# Funcion que nos da la informacion de los dataframes.
def informacion(df, nombre=''):
  print(f'\n========================================')
  print(nombre.center(40, '='))
  print(f'========================================\n')
  print(f'\nEstructura de la tabla:\n {df.shape}\n')
  print(f'\nTipos de datos\n {df.dtypes}\n')
  print(df.head())
  print('\n')
  print(df.tail())
  print(f'\nResumen estadistico prematuro con RawData:\n {df.describe()}\n')
```

```python
informacion(df_marketing, 'Marketing')
informacion(df_clientes, 'Clientes')
informacion(df_ventas, 'Ventas')
```

```python
# 5. Calidad de Datos: Identificar valores nulos y duplicados en los conjuntos de
# datos. Documentar el estado inicial de los datos.

# Contamos duplicados por fila completa en el DataFrame

def filas_duplicadas(df, nombre=''):
  print(f'\n========================================')
  print(nombre.center(40, '='))
  print(f'========================================\n')

  duplicados = df.duplicated(keep=False)
  total_duplicados = duplicados.sum()
  print(f'Filas totales: {len(df)}')
  print(f'Total de filas duplicadas: {total_duplicados}')
  print(f'Filas sin duplicar: {len(df) - total_duplicados}\n')

# Verificamos y contamos datos Nulos.

def filas_nulas(df, nombre=''):
  nulos = df.isnull()
  total_nulos = nulos.sum()
  print(f'Total de filas nulas:\n{total_nulos}\n')
  print(f'Filas sin duplicar:\n{len(df) - total_nulos}\n')

# Verificamos y sumamos ya si existe algun (any) nulo en el Data Frame.

def datos_nulos(df, nombre=''):
    if df.isnull().values.any():
        print(f'El Data Fame {nombre.upper()} contiene:\n{df.isnull().sum()}\n...datos nulos...')
    else:
        print(f'El Data Frame {nombre.upper()} no contiene nulos entre sus datos.')

```

```python
filas_duplicadas(df_marketing, 'Marketing')
filas_nulas(df_marketing, 'Marketing')
datos_nulos(df_marketing, 'Marketing')
filas_duplicadas(df_clientes, 'Clientes')
filas_nulas(df_clientes, 'Clientes')
datos_nulos(df_clientes, 'Clientes')
filas_duplicadas(df_ventas, 'Ventas')
filas_nulas(df_ventas, 'Ventas')
datos_nulos(df_ventas, 'Ventas')
```

```python
# 6- Limpieza de Datos: Limpiar el conjunto de datos eliminando
# duplicados y caracteres no deseados.
# Documentar el proceso y los resultados.
# Pasos de limpieza:
# A- Eliminar duplicados.
# B- Eliminar caracteres no deseados.
# C- Normalizar String y datos DATE.
# D- Rellenar datos nulos.


marketing_limpio = df_marketing.copy()
clientes_limpio = df_clientes.copy()
ventas_limpio = df_ventas.copy()

# A- Eliminar duplicados.
marketing_limpio = marketing_limpio.drop_duplicates()
clientes_limpio = clientes_limpio.drop_duplicates()
ventas_limpio = ventas_limpio.drop_duplicates()
```

```python
# B- Eliminamos caracteres y transformamos los datos de VENTAS(precio) a float.

def limpiar_precios(df, columna='precio'):

    if columna in df.columns:
      df[columna] = (df[columna]
                     .astype(str)
                     .str.strip()
                     .str.replace(' ', '')
                     .str.replace('$', '', regex=False)
                     .str.replace(',', ''))
      df[columna] = pd.to_numeric(df[columna], errors='coerce')
      nulos = df[columna].isna().sum()

      print(f'\nLa cantidad de datos tipo NaN es de = {nulos}.\n')
      print(f'\nLa columna {columna} se pudo convertir al Tipo de dato: {df[columna].dtype} con exito.\n')

      return df

    else:
      print(f'\n====La columna {columna} no existe en el Data Frame.====\n')
```

```python
limpiar_precios(ventas_limpio,'precio')
print(ventas_limpio)
```

```python
# C- Limpiar string de los Data Frame normalizar y
# cambiar los necesarios a DATE

def limpiar_fechas(df):

   for columna in df.columns:
    if 'fecha' in columna.lower():
      df[columna] = pd.to_datetime(df[columna], errors='coerce', dayfirst=True)

   # Opcional: analizar cuántos valores se convirtieron a NaT
   no_time = df[columna].isna().sum()
   if no_time > 0:
    print(f'El Data Frame {df.upper()} tiene {no_time} fechas inválidas convertidas a NaT')

    return df
```

```python
limpiar_fechas(marketing_limpio)
limpiar_fechas(ventas_limpio)
print(marketing_limpio)
print(ventas_limpio)
```

```python
# D- Rellenar datos nulos y NaN con el promedio.
# Para cantidad rellenamos con 0 (cero)
######################  EL UNICO DATA FRAME CON NULOS ES VENTAS #################
# Aplicaremos esta funcion solo en él.

def rellenar_nulos(df, columna='cantidad'):
  # columna CANTIDAD
  if 'cantidad' in df.columns:
    nulos_cantidad = df['cantidad'].isnull().sum() #suma los nulos
    print(f'Cantidad de campos nulos en CANTIDAD: {nulos_cantidad}.\n')
    if nulos_cantidad > 0:
      df['cantidad'].fillna(0) #los reemplaza
      print(f'{nulos_cantidad} campos CANTIDAD nulos pasados a 0 (cero).\n')
    else:
      print('No hay campos nulos en CANTIDAD.\n')
  return df


def rellenar_Nan(df, columna='precio'):
  # columna PRECIO
  if 'precio' in df.columns:
    na_precio = df['precio'].isna().sum()
    print(f'Cantidad de campos de tipo NaN en PRECIO: {na_precio}.\n')
    if na_precio > 0:
      precio_promedio = df['precio'].mean()
      df['precio'].fillna(precio_promedio, inplace=True)
      print(f'{na_precio} campos de tipo NaN en PRECIO pasados a {precio_promedio:.2f} (promedio).\n')
    else:
      print('No hay campos de tipo NaN en PRECIO.\n')

  return df
```

```python
rellenar_nulos(ventas_limpio, 'cantidad')
rellenar_Nan(ventas_limpio, 'precio')
print(ventas_limpio)
```

```python
# ===================================================================
# ============== INFORME COMPARATIVO DE LOS DATA FRAME ==============
# ===================================================================
# Compararemos transformacion de datos, nulos, filas columnas, duplicados.

def informe_comparativo(df_original, df_limpio, nombre_original='', nombre_limpio=''):

    informe_datos = [
        ["Archivo", nombre_original, nombre_limpio],
        ["Filas", f"{df_original.shape[0]}", f"{df_limpio.shape[0]}"],
        ["Columnas", df_original.shape[1], df_limpio.shape[1]],
        ["Nulos", f"{df_original.isnull().sum().sum()}", f"{df_limpio.isnull().sum().sum()}"],
        ["Duplicados", f"{df_original.duplicated().sum()}", f"{df_limpio.duplicated().sum()}"]
    ]

    cuadro = tabulate(
        informe_datos,
        headers=['', 'ORIGINAL', 'LIMPIO'],
        tablefmt='grid',
        stralign='center'
    )

    return print(cuadro)
```

```python
informe_comparativo(df_marketing, marketing_limpio, 'marketing', 'marketing_limpio')
informe_comparativo(df_clientes, clientes_limpio, 'clientes', 'clientes_limpio')
informe_comparativo(df_ventas, ventas_limpio, 'ventas', 'ventas_limpio')
```
